```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('/content/Spotify Tracks Dataset dataset.csv')
```

```python
df.head()
```

```python
df['track_genre'].unique()
```

```python
df = df[
    df['track_genre'].isin(
        ['club', 'pop', 'brazil', 'blues', 'brazil', 'british']
    )
    ]
```

```python
df.head()
```

```python
df.drop(
    columns=[
        'Unnamed: 0',
        'track_id',
        'track_name',
        'artists',
        'album_name',
        'time_signature'
    ],
    axis=1,
    inplace = True
)
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
x = df.drop('track_genre', axis=1)
y = df['track_genre']
```

```python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

Label output

```python
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_val = label_encoder.transform(y_val)
```

Scaling features

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
```

**Model**

```python
xgb_classifier = xgb.XGBClassifier(
    random_state=42
)

xgb_classifier.fit(x_train, y_train)
```

```python
y_pred = xgb_classifier.predict(x_val)
```

```python
val_accuracy = accuracy_score(y_val, y_pred)
print(f"Validation Accuracy: {val_accuracy: .4f}")
```

```python
print(classification_report(y_val, y_pred))
```

```python
print(confusion_matrix(y_val, y_pred))
```

Hyperparameter tuning

```python
param_grid = {
    'n_estimators':[100, 200, 300, 400],
    'learning_rate':[0.01, 0.1, 0.2],
    'max_depth':[3, 6, 9],
    'min_child_weight':[1, 3, 5],
    'subsample':[0.7, 0.85, 1],
    'colsample_bytree':[0.7, 0.85, 1],
    'reg_alpha':[1, 3, 5],
    'reg_lambda':[1]
}
grid_search = RandomizedSearchCV(
    estimator=xgb.XGBClassifier(
        random_state=42,
        eval_metric='mlogloss'
    ),
    param_distributions=param_grid,
    cv=5,
    scoring='accuracy',
    n_iter=20,
    n_jobs=-1,
    random_state=42
)
```

```python
grid_search.fit(x_train, y_train)
best_xgb = grid_search.best_estimator_
print('Best parameters:')
print(grid_search.best_params_)
print('\nBest cv score:', grid_search.best_score_)
```

```python
y_pred = best_xgb.predict(x_val)

test_accuracy = accuracy_score(y_val, y_pred)

print("Test Accuracy:", test_accuracy)

print("\nClassification Report:")
print(classification_report(y_val, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_val, y_pred))
```
